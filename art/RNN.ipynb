```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
text = "I am an IT Student."
chars = sorted(list(set(text)))
chars_to_index = {char: i for i, char in enumerate(chars)}
index_to_char = {i: char for i, char in enumerate(chars)}
```

```python
seq_length = 3
sequences = []
labels = []

for i in range(len(text) - seq_length):
  seq = text[i: i + seq_length]
  label = text[i + seq_length]

  sequences.append([chars_to_index[char] for char in seq])
  labels.append(chars_to_index[label])

x = np.array(sequences)
y = np.array(labels)
```

```python
x_one_hot = tf.one_hot(x, len(chars))
y_one_hot = tf.one_hot(y, len(chars))
```

```python
model = Sequential()
model.add(SimpleRNN(50, input_shape = (seq_length, len(chars)), activation = "relu"))
model.add(Dense(len(chars), activation = "softmax"))
```

```python
model.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])
model.fit(x_one_hot, y_one_hot, epochs = 100)
```

```python
start_seq = "I am S"
genrated_text = start_seq
for i in range(50):
  x_seq = [chars_to_index[char] for char in genrated_text[-seq_length:]]
  x_seq_one_hot = tf.one_hot(x_seq, len(chars))
  x_seq_one_hot = tf.expand_dims(x_seq_one_hot, axis = 0)
  pred = model.predict(x_seq_one_hot)
  pred_char = index_to_char[np.argmax(pred)]
  genrated_text += pred_char

print(genrated_text)
```
