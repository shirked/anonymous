```python
def menu():
  print("---------Fuzzy Operations----------------")
  print("[0] Declare new A, B sets")
  print("[1] Union")
  print("[2] Intersection")
  print("[3] Complement A")
  print("[4] Complement B")
  print("[5] Algebraic Sum")
  print("[6] Algebraic Product")
  print("[7] Bounded sum of A and B")
  print("[8] Bounded Difference of A and B")
  print("[9] Quit")
  print("")
```

```python
def newInputs():
  global SetA
  global SetB
  del SetA[:]
  del SetB[:]
  SetA = []
  SetB = []

  print("----Set A-----")
  n = int(input("Enter the number of values:-"))
  for i in range(n):
    n, d = map(float, input(f"[Set A Input - {i + 1}] Enter values (0.1/1) -> (0.1/1): \n").split())
    SetA.append([n, d])

  print("----Set B----")
  n = int(input("Enter the number of values:-"))
  for i in range(n):
    n, d = map(float, input(f"[SetB Input - {i + 1}] Enter values (0.1/1) -> (0.1/1): \n").split())
    SetB.append([n, d])

  print("Set A:- ", SetA)
  print("Set B:- ", SetB)
```

```python
def checkElementHelper(x, S):
  for e in S:
    if x == e[1]:
      return e[0]
```

```python
from copy import deepcopy
```

```python
def union(SetA, SetB):
  X, Y = deepcopy(SetA), deepcopy(SetB)
  Z = []

  for i in X:
    mb = checkElementHelper(i[1], Y)
    Z.append([max(mb, i[0]), i[1]])
    if mb != 0:
      Y.remove([mb, i[1]])
  Z = Z + Y
  return Z
```

```python
def intersection(SetA, SetB):
  X, Y = deepcopy(SetA), deepcopy(SetB)
  Z = []

  for i in X:
    mb = checkElementHelper(i[1], Y)
    if min(mb, i[0]) != 0:
      Z.append([min(mb, i[0]), i[1]])

  return Z
```

```python
def complement(setA):
  Z = deepcopy(setA)
  for i in Z:
    i[0] = 1 - i[0]
    if i[0] == 0:
      Z.remove(i)
  return Z
```

```python
from ast import Yield
def algebraicSum(setA, setB):
  X, Y = deepcopy(setA), deepcopy(setB)
  Z = []

  for i in X:
    mb = checkElementHelper(i[1], Y)
    if mb != 0:
      m = round(i[0] + mb - (i[0] * mb), 3)
      Z.append([m, i[1]])
      Y.remove([mb, i[1]])
    else:
      Z.append(i)
  Z = Z + Y
  return Z
```

```python
def algebraicProduct(setA, setB):
  X, Y = deepcopy(setA), deepcopy(setB)
  Z = []

  for i in X:
    mb = checkElementHelper(i[1], Y)
    if mb != 0:
      m = round(i[0] * mb, 3)
      Z.append([m, i[1]])
      Y.remove([mb, i[1]])
    else:
      Z.append(i)
  return Z
```

```python
def boundedSum(setA, setB):
  X, Y = deepcopy(setA), deepcopy(setB)
  Z = []

  for i in X:
    mb = checkElementHelper(i[1], Y)
    if mb != 0:
      m = i[0] + mb
      Z.append([min((1, m)), i[1]])
      Y.remove([mb, i[1]])
    else:
      Z.append(i)
  Z = Z + Y
  return Z
```

```python
def boundedDifference(setA, setB):
  X, Y = deepcopy(setA), deepcopy(setB)
  Z = []

  for i in X:
    mb = checkElementHelper(i[1], Y)
    if mb != 0:
      m = i[0] - mb
      Z.append([max(0, m), i[1]])
      Y.remove([mb, i[1]])
    else:
      Z.append(i)
  return Z
```

```python
def fuzzyOperate(c):
  if c == 0:
    print("New sets:-")
    newInputs()
  elif c == 1:
    print("Union:-")
    print(union(SetA, SetB))
  elif c == 2:
    print("Intersection:-")
    print(intersection(SetA, SetB))
  elif c == 3:
    print("Complement of A:-")
    print(complement(SetA))
  elif c == 4:
    print("Complement of B:-")
    print(complement(SetB))
  elif c == 5:
    print("Algebraic Sum:-")
    print(algebraicSum(SetA, SetB))
  elif c == 6:
    print("Algebraic Product:-")
    print(algebraicProduct(SetA, SetB))
  elif c == 7:
    print("Bounded Sum:-")
    print(boundedSum(SetA, SetB))
  elif c == 8:
    print("Bounded Difference:-")
    print(boundedDifference(SetA, SetB))
  else:
    print("Invalid choice.")
```
